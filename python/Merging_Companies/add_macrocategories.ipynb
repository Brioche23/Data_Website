```python
import pandas as pd

# Step 1: Read the CSV file into a pandas DataFrame
df = pd.read_csv('datasets/breaches_exp_geo.csv')

df.head(3)
```

```python

# Step 2: Clean the 'Organization Type' column by removing square brackets and single quotes
df['Organization Type'] = df['Organization Type'].str.strip("[]'")

df.head(3)
```

```python
# Step 3: Create a mapping for the macrocategories
macrocategory_mapping = {
    'healthcare': 'Health & Science',
    'biotechnology': 'Health & Science',
    'academic': 'Health & Science',
    'educational': 'Health & Science',
    'financial': 'Finance & Business',
    'retail': 'Finance & Business',
    'online marketing': 'Finance & Business',
    'data broker': 'Finance & Business',
    'tech': 'Technology & Telecommunications',
    'telecoms': 'Technology & Telecommunications',
    'hosting provider': 'Technology & Telecommunications',
    'government': 'Government & Public Services',
    'military': 'Government & Public Services',
    'humanitarian': 'Government & Public Services',
    'political': 'Government & Public Services',
    'charity': 'Government & Public Services',
    'social network': 'Media & Entertainment',
    'media': 'Media & Entertainment',
    'entertainment': 'Media & Entertainment',
    'web': 'Media & Entertainment',
    'gaming': 'Media & Entertainment',
    'advertising': 'Media & Entertainment',
    'publishing': 'Media & Entertainment',
    'construction': 'Industry & Infrastructure',
    'energy': 'Industry & Infrastructure',
    'transport': 'Industry & Infrastructure',
    'hospitality': 'Hospitality & Services',
    'restaurant': 'Hospitality & Services',
    'local search': 'Hospitality & Services',
    'various': 'Other'
}

# Step 4: Map the cleaned 'Organization Type' values to the macrocategories
df['macrocategory'] = df['Organization Type'].map(macrocategory_mapping)


df.head(5)
```

```python

# Step 6: Save the result to a new CSV file
df.to_csv('breaches_exp_geo_macro.csv', index=False)
```
