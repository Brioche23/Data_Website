```python

# Directory to save image files
output_dir = 'logos'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Iterate over each company
for index, row in df.iterrows():
    company_name = row['Company']
    img_url = fetch_first_image_url(company_name)
    print(f"Response: {img_url}")
    
    if img_url:
        try:
            img_response = requests.get(img_url)
            img_response.raise_for_status()  # Raise HTTPError for bad responses
            file_extension = img_url.split('.')[-1].split('?')[0]  # Extract file extension
            if not file_extension:
                file_extension = 'png'  # Default to PNG if extension is missing
            file_path = os.path.join(output_dir, f"{company_name}.{file_extension}")
            with open(file_path, 'wb') as file:
                file.write(img_response.content)
            print(f"Downloaded logo for {company_name}")
        except requests.exceptions.RequestException as e:
            print(f"Failed to download logo for {company_name}: {e}")
    else:
        print(f"No logo found for {company_name}")

print("Completed downloading logos.")
```
