```python
# Define a function to convert abbreviated numbers to integers
def convert_to_int(abbrev_num):
    if isinstance(abbrev_num, str):
        if 'k' in abbrev_num:
            return int(float(abbrev_num.replace('k', '')) * 1000)
        elif 'M' in abbrev_num:
            return int(float(abbrev_num.replace('M', '')) * 1000000)
        elif 'B' in abbrev_num:
            return int(float(abbrev_num.replace('B', '')) * 1000000000)
        else:
            return int(abbrev_num)
    else:
        return abbrev_num

# Apply the function to every cell in the DataFrame except for the 'Country' column
for column in itUsers.columns:
    if column != 'country':
        itUsers[column] = itUsers[column].apply(convert_to_int)
for column in nonItUsers.columns:
    if column != 'country':
        nonItUsers[column] = nonItUsers[column].apply(convert_to_int)

itUsers.head(3)

```

```python
# Reshape the DataFrame so that each country has multiple rows, one for each year
itUsers = itUsers.melt(id_vars=['country'], var_name='year', value_name='Users')
nonItUsers = nonItUsers.melt(id_vars=['country'], var_name='year', value_name='Non Users')

# Optionally, you can sort the DataFrame by 'country' and 'Year'
itUsers = itUsers.sort_values(by=['country', 'year']).reset_index(drop=True)
nonItUsers = nonItUsers.sort_values(by=['country', 'year']).reset_index(drop=True)
```

```python
nonItUsers.head(30)
```

```python
merged = pd.merge(itUsers, nonItUsers, on = ["country","year"], how="outer").fillna(0)
```

```python
merged.head(3)
```

```python
filteredITA = merged[merged["country"] == "Italy"].reset_index(drop=True)
filteredITA.sort_values("year")
filteredITA.head(20)

```

```python
usersITA = filteredITA[filteredITA["Users"]!= 0].reset_index(drop=True)
nonUsersITA = filteredITA[filteredITA["Non Users"]!= 0].reset_index(drop=True)
# plt.plot(usersITA["year"], usersITA["Users"], marker = 'o')
# plt.bar(usersITA["year"], usersITA["Users"])
# plt.bar(nonUsersITA["year"], nonUsersITA["Non Users"])

fig, ax = plt.subplots()
bottom = np.zeros(3)    

ax.stackplot(usersITA["year"], nonUsersITA["Non Users"], usersITA["Users"] )
ax.set_title("Internet Users vs Non Users in Italian Population")
labels=["Non Users", "Users"]
ax.legend(labels, loc="lower left")
plt.xticks([1, 10, 20, 29])
ax.set_xlabel('Year')
ax.set_ylabel('Population in 10M')
ax.grid(True, alpha=.2)

# for boolean, weight_count in weight_counts.items():
#     p = ax.bar(species, weight_count, width, label=boolean, bottom=bottom)
#     bottom += weight_count
```
